```python

 #nltk.download()
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from textblob import TextBlob

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')

# Initialize lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Function to preprocess text
def preprocess_text(text):
    tokens = word_tokenize(text.lower())
    tokens = [lemmatizer.lemmatize(token) for token in tokens if token.isalnum()]
    tokens = [token for token in tokens if token not in stop_words]
    return tokens

# Apply preprocessing to the 'Text' column
df_text_engl['Processed_Text'] = df_text_engl['Text'].apply(preprocess_text)
print(df_text_engl.head())
```

```python
# Tokenization and word count
df_text_engl['Word_Count'] = df_text_engl['Processed_Text'].apply(len)

# Sentiment Analysis
df_text_engl['Sentiment'] = df_text_engl['Text'].apply(lambda x: TextBlob(x).sentiment.polarity)

# Most common words
from collections import Counter

def get_top_words(text_list, n=10):
    word_freq = Counter([word for sublist in text_list for word in sublist])
    return word_freq.most_common(n)

top_words = get_top_words(df_text_engl['Processed_Text'])

# Named Entity Recognition (NER)
import spacy

nlp = spacy.load("en_core_web_sm")

def extract_entities(text):
    doc = nlp(text)
    return [(ent.text, ent.label_) for ent in doc.ents]

df_text_engl['Entities'] = df_text_engl['Text'].apply(extract_entities)

# Topic Modeling
from gensim import corpora
from gensim.models.ldamodel import LdaModel

dictionary = corpora.Dictionary(df_text_engl['Processed_Text'])
corpus = [dictionary.doc2bow(text) for text in df_text_engl['Processed_Text']]

lda_model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=10, random_state=100)

# Print results
print("Average Word Count:", df_text_engl['Word_Count'].mean())
print("\nAverage Sentiment:", df_text_engl['Sentiment'].mean())
print("\nTop 10 Most Common Words:", top_words)
print("\nSample Named Entities:", df_text_engl['Entities'].iloc[0][:5])
print("\nTop 5 Topics:")
for idx, topic in lda_model.print_topics(-1):
    print(f"Topic {idx}: {topic}")
```

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.sentiment import SentimentIntensityAnalyzer
from collections import Counter, defaultdict
import re

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('vader_lexicon')
nltk.download('punkt_tab')


# Initialize stopwords and sentiment analyzer
stop_words = set(stopwords.words('english'))
sia = SentimentIntensityAnalyzer()

def preprocess_text(text):
    # Remove special characters and numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Convert to lowercase
    text = text.lower()
    return text


# Function to preprocess text
def preprocess_text(text):
    tokens = word_tokenize(text.lower())
    tokens = [token for token in tokens if token.isalnum() and token not in stop_words]
    return tokens

# Function to get sentiment of words
def get_word_sentiment(word):
    return sia.polarity_scores(word)['compound']

# Analyze sentiment for each text entry
sentiment_analysis_word = []
overall_positive_count = defaultdict(int)
overall_negative_count = defaultdict(int)

for text in df_text_engl['Text']:
    processed_text = preprocess_text(text)

    # Get sentiment for each word
    word_sentiments = {word: get_word_sentiment(word) for word in set(processed_text)}

    # Separate positive and negative words
    positive_words = [word for word, sentiment in word_sentiments.items() if sentiment > 0]
    negative_words = [word for word, sentiment in word_sentiments.items() if sentiment < 0]

    # Update overall counts
    for word in positive_words:
        overall_positive_count[word] += 1
    for word in negative_words:
        overall_negative_count[word] += 1

    # Store results
    sentiment_analysis_word.append({
        'Positive_Words': Counter(positive_words).most_common(10),
        'Negative_Words': Counter(negative_words).most_common(10)
    })

# Add sentiment analysis results to the dataframe
df_text_engl['Sentiment_Analysis_word'] = sentiment_analysis_word

# Print results for each text entry
for index, row in df_text_engl.iterrows():
    print(f"\nRow {index + 1}:")
    print("Brand:", row['Brand'])
    print("Language:", row['Language'])
    print("Summary:", row['Text'][:100] + "..." if len(row['Text']) > 100 else row['Text'])
    print("Top 5 Positive Words:", row['Sentiment_Analysis']['Positive_Words'])
    print("Top 5 Negative Words:", row['Sentiment_Analysis']['Negative_Words'])

# Get overall top ten positive and negative words
overall_top_positive = Counter(overall_positive_count).most_common(10)
overall_top_negative = Counter(overall_negative_count).most_common(10)

print("\nOverall Top 10 Positive Words:", overall_top_positive)
print("Overall Top 10 Negative Words:", overall_top_negative)
```
